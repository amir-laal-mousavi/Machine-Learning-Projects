```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3,4,6,9,10,15,13])
Y = np.array([2,6,8,11,15,21,30,40,34])
print(X.shape)
print(Y.shape)
```

```python
def plotData(x, y):
    plt.plot(x, y, 'ro', ms=10, mec='k')
    plt.ylabel('eggs')
    plt.xlabel('chicken')
```

```python
def abline(slope, intercept, epoch):
  axes = plt.gca()
  x_vals = np.array(axes.set_xlim(-20,20))
  y_vals = intercept + slope * x_vals
  plt.ylim((-50,50))
  plt.figure(epoch)
  plt.plot(x_vals,y_vals, '--')
  plt.text(0,-4,'epoch' + str(epoch + 1))
```

```python
plotData(X,Y)
```

```python
# Random initializing theta0 and theta1
theta0 = 1
theta1 = 0
# Hyper parameters and training
lr = 0.025
m = X.shape[0]
# How many times are we gonna repeat the process?
epochs = 20
for epoch in range(epochs):
  sum0 = 0
  sum1 = 0
  for i in range(m):
    sum0 = sum0 + (theta0 + theta1 * X[i] - Y[i])
    sum1 = sum1 + (theta0 + theta1 * X[i] - Y[i]) * X[i]

  theta0 = theta0 - (lr/m) * (sum0)
  theta1 = theta1 - (lr/m) * (sum1)

  plotData(X, Y)
  abline(theta1, theta0, epoch)
```
